```python
import matplotlib.pyplot as mp
```

```python
x=[1,2,3,4,5,6]
y=[1,2,3,4,5,6]
```

```python
mp.plot(x,y)
```

```python
y1=[1.5,2.5,3.5,4.5,5.5,6.5]
mp.plot(x,y1)
```

```python
mp.plot(x,y)
mp.plot(x,y1)
```

```python
mp.plot(x,y,label='1 dateset')
mp.plot(x,y1,label='2 dataset')
```

```python
mp.plot(x,y,label='1 dateset')
mp.plot(x,y1,label='2 dataset')
mp.legend()
```

```python
mp.plot(x,y,label='1 dateset')
mp.plot(x,y1,label='2 dataset')
mp.legend()
mp.show()
```

```python
mp.plot(x,y,label='1 dateset')
mp.plot(x,y1,label='2 dataset')
mp.legend()
mp.title("Comparing 2 trands")
mp.show()
```

```python
mp.plot(x,y,label='1 dateset')
mp.plot(x,y1,label='2 dataset')
mp.legend()
mp.title("Comparing 2 Trands")
mp.xlabel("Rate of Inflation")
mp.ylabel("Cost of Living")
mp.show()
```

```python
mp.plot(x,y,label='1 dateset',color='g')
mp.plot(x,y1,label='2 dataset',color='r')
mp.legend()
mp.title("Comparing 2 Trands")
mp.xlabel("Rate of Inflation")
mp.ylabel("Cost of Living")
mp.show()
```

```python
mp.plot(x,y,label='1 dateset',color='g',marker='o')
mp.plot(x,y1,label='2 dataset',color='r',marker='^')
mp.legend()
mp.title("Comparing 2 Trands")
mp.xlabel("Rate of Inflation")
mp.ylabel("Cost of Living")
mp.show()
```

```python
fig,(ax1,ax2)=mp.subplots(1,2)
ax1.plot(x,y,label='1 dateset',color='g',marker='o')
ax2.plot(x,y1,label='2 dataset',color='r',marker='^')
```
