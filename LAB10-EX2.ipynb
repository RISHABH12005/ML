```python
import numpy as n
import matplotlib.pyplot as mp
```

```python
n.random.seed(42)
m=100
x=2*n.random.rand(m,1)
y=4+3*x+n.random.rand(m,1)
```

```python
x[:10],y[:10]
```

```python
mp.plot(x,y,"b.")
```

```python
from sklearn.preprocessing import add_dummy_feature
x_b=add_dummy_feature(x)
theta_best=n.linalg.inv(x_b.T@x_b)@x_b.T@y
```

```python
theta_best
```

```python
test_data=n.array([[0],[2]])
test_data_b=add_dummy_feature(test_data)
y_predict=test_data_b@theta_best
y_predict
```

```python
test_data
```

```python
mp.plot(x,y,"b.")
mp.plot(test_data,y_predict,"ro")
mp.plot(test_data,y_predict,"g--")
```
