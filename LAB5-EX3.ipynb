```python
import matplotlib.pyplot as mp
import numpy as n
import pandas as p
```

```python
mp.pie([1])
mp.show()
```

```python
x=[2,23,14,5,6]
mp.pie(x,labels=x)
mp.show()
```

```python
y=["Defense","Education","Health","Infra","Agri"]
mp.pie(x,labels=y)
mp.show()
```

```python
mp.pie(x,labels=y,counterclock=False)
mp.show()
```

```python
mp.pie(x,labels=y,explode=[0.15,0,0,0,0])
mp.show()
```

```python
mp.pie(x,labels=y,explode=[0.15,0,0,10.0,0])
mp.show()
```

```python
mp.pie(x,labels=y,colors=['r','g','b','y','m'])
mp.show()
```

```python
mp.pie(x,labels=y)
mp.show()
```

```python
mp.pie(x,labels=y,rotatelabels=True)
mp.show()
```

```python
mp.pie(x,labels=y,frame=True)
mp.show()
```

```python
mp.pie(x,labels=y,autopct="%0.2f%%",shadow=True)
mp.show()
```

```python
mp.pie(x,labels=y,autopct="%0.2f%%",radius=1.5)
mp.show()
```

```python
mp.pie(x,labels=y,autopct="%0.2f%%",radius=1.5,labeldistance=0.8)
mp.show()
```

```python
mp.pie(x,labels=y,labeldistance=0.6,startangle=90)
mp.show()
```

```python
mp.pie(x,labels=y,autopct="%0.2f%%",radius=1.5,labeldistance=0.8,textprops={"fontsize":15,'color':'y'})
mp.show()
```

```python
mp.pie(x,labels=y,explode=[0.2,0,0,0,0],textprops={'fontsize':15},wedgeprops={'width':1.2})
mp.show()
```

```python
mp.pie(x,labels=y,explode=[0.2,0,0,0,0],textprops={'fontsize':15},wedgeprops={'linewidth':2,'edgecolor':'y'})
mp.show()
```

```python
mp.pie(x,labels=y)
mp.title("Budget Distubition")
mp.legend()
mp.show()
```
