```python
import pandas as p
import numpy as n
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import RidgeCV
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error
```

```python
a=p.read_csv('housing.csv')
a
```

```python
a.info()
```

```python
a.head()
```

```python
x=a.drop("median_house_value",axis=1)
y=a["median_house_value"]
```

```python
numf =x.select_dtypes(include=["float64"]).columns
catf =x.select_dtypes(include=["object"]).columns
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
nump=Pipeline([("imputer",SimpleImputer(strategy="median")),("scaler",StandardScaler())])
catp=Pipeline([("encoder",OneHotEncoder(handle_unknown="ignore"))])
pp=ColumnTransformer([("num",nump,numf),("cat",catp,catf)])
model=Pipeline([("preprocessor",pp),("regressor",RidgeCV(alphas=[0.1,1.0,10.0],cv=5))])
```

```python
model.fit(x_train,y_train)
```

```python
trainp=model.predict(x_train)
testp=model.predict(x_test)
```

```python
trainr=n.sqrt(mean_squared_error(y_train,trainp))
testr=n.sqrt(mean_squared_error(y_test,testp))
```

```python
print("Train RMSE:",trainr)
print("Test RMSE:",testr)
```
