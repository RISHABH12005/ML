```python
import numpy as n
import pandas as p
import matplotlib.pyplot as mp
```

```python
mp.hist([12,8,15,8,15,8,5,20])
```

```python
x=n.random.randint(10,60,(50))
```

```python
mp.hist(x,color='g',edgecolor='r')
mp.title("Histogram Demo")
mp.xlabel('City Code')
mp.ylabel('Population')
```

```python
mp.hist(x,color='g',bins=[10,20,30,40,50,60],edgecolor='r')
```

```python
len(x[x<20]),len(x[(x>=20)&(x<30)])
```

```python
mp.hist(x,"auto",edgecolor='r',color='g')
```

```python
mp.hist(x,range=(0,100),edgecolor='r',color='g')
```

```python
mp.hist(x,"auto",edgecolor='r',color='g',cumulative=-1)
```

```python
mp.hist(x,"auto",edgecolor='r',color='g',bottom=20)
```

```python
mp.hist(x,"auto",edgecolor='r',color='g',bottom=20,align='mid')
```

```python
mp.hist(x,"auto",edgecolor='r',color='g',bottom=20,align='mid',histtype='stepfilled')
```

```python
mp.hist(x,"auto",edgecolor='r',color='g',orientation='horizontal')
```

```python
mp.hist(x,"auto",edgecolor='r',color='g',rwidth=0.7)
```

```python
mp.hist(x,"auto",edgecolor='r',color='g',rwidth=0.7,log=True)
```

```python
mp.hist(x,"auto",edgecolor='r',color='g',rwidth=0.7,label="City-wise Population")
mp.legend()
```

```python
mp.hist(x,"auto",color='g',rwidth=0.7,label="City-wise Population")
mp.axvline(35,color='r',label="break point")
mp.legend()
```
